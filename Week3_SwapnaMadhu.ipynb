```python
# Quesion 3
# Which is the product where u have seen decrease in price
'''
This can be answered by grouping part_number and description which gives unique products and 
respective price and take difference b/w the values. If the diff is positive which indicates the
price is decreased.
'''
df1.groupby(['part_number','description'])["price"].diff().sort_values(ascending=False).head(10)
```

```python
# Quesion 4
# Which is the product where u have seen increase in price
'''
This can be answered by grouping part_number and description which gives unique products and 
respective price and take difference b/w the values. If the diff is neagtive which indicates the
price is increased.
'''
df1.groupby(['part_number','description'])["price"].diff().sort_values().head()
```

```python
# Question 6
# Which products are aging
'''
This can be determined by getting unique values by grouping using 
sku_create_date,price,category,part_number and description and sorting the values and 
top values give the products which are aging because they are in stock for long period
'''
age_df1=df1.groupby(['sku_create_date','price','category','part_number','description']).sum()
top_age_df1=age_df1.sort_values(['qty'],ascending=False)
top_age_df1.head(20)
```
